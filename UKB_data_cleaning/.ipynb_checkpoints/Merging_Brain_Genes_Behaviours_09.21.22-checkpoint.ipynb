```python
from sklearn.cross_decomposition import CCA
import pandas as pd
import joblib
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
```

# Importing CCA modes

```python
cca_ids = pd.read_csv('org_CCA_k25_lonely_full.csv', index_col=0, header=None)
cca_ids.columns = ['eid']
pop_cca = joblib.load('org_CCA_k25_lonely_full')
#x_scores:CCA HP, y_scores: CCA DMN
cca_ids = pd.DataFrame(
#stacking data in a single numpy array    
    np.vstack((
        cca_ids.values.T, pop_cca.x_scores_.T, pop_cca.y_scores_.T))).T
#adding the eids column
cca_ids.columns = ['eid'] + list(np.arange(0,50)+1)
cca_ids['eid']= cca_ids['eid'].astype(np.int64)
```

# **Aligning CCA with APOE**

```python
import pandas as pd
import os 

#'eid', 'rs429358-1', 'rs429358-2', 'rs7412-1', 'rs7412-2', 'APOE-1', 'APOE-2'
def main():
    df = pd.read_csv('apoe.csv')
    df['Allele-1'] = False
    df['Allele-2'] = False
    df['Genotypes'] = False

    print(df.shape)
    print(df.columns)
    print(df)
    # APOE 1
    df.loc[(df['rs429358-1'] == 'C') & (df['rs7412-1'] == 'T'), 'Allele-1'] = 'APOE1'
    df.loc[(df['rs429358-2'] == 'C') & (df['rs7412-2'] == 'T'), 'Allele-2'] = 'APOE1'
    # APOE 2
    df.loc[(df['rs429358-1'] == 'T') & (df['rs7412-1'] == 'T'), 'Allele-1'] = 'APOE2'
    df.loc[(df['rs429358-2'] == 'T') & (df['rs7412-2'] == 'T'), 'Allele-2'] = 'APOE2'
    # APOE 3
    df.loc[(df['rs429358-1'] == 'T') & (df['rs7412-1'] == 'C'), 'Allele-1'] = 'APOE3'
    df.loc[(df['rs429358-2'] == 'T') & (df['rs7412-2'] == 'C'), 'Allele-2'] = 'APOE3'
    # APOE 4	
    df.loc[(df['rs429358-1'] == 'C') & (df['rs7412-1'] == 'C'), 'Allele-1'] = 'APOE4'
    df.loc[(df['rs429358-2'] == 'C') & (df['rs7412-2'] == 'C'), 'Allele-2'] = 'APOE4'

    #Genotypes
    #homozygous
    df.loc[(df['Allele-1']=='APOE1') & (df['Allele-2']=='APOE1'), 'Genotypes'] = 'e1/e1'
    df.loc[(df['Allele-1']=='APOE2') & (df['Allele-2']=='APOE2'), 'Genotypes'] = 'e2/e2'
    df.loc[(df['Allele-1']=='APOE3') & (df['Allele-2']=='APOE3'), 'Genotypes'] = 'e3/e3'
    df.loc[(df['Allele-1']=='APOE4') & (df['Allele-2']=='APOE4'), 'Genotypes'] = 'e4/e4'
    #heterozygous
    #E1/E2
    df.loc[(df['Allele-1']=='APOE1') & (df['Allele-2']=='APOE2'), 'Genotypes'] = 'e1/e2'
    df.loc[(df['Allele-1']=='APOE2') & (df['Allele-2']=='APOE1'), 'Genotypes'] = 'e1/e2'
    #E1/E3
    df.loc[(df['Allele-1']=='APOE1') & (df['Allele-2']=='APOE3'), 'Genotypes'] = 'e1/e3'
    df.loc[(df['Allele-1']=='APOE3') & (df['Allele-2']=='APOE1'), 'Genotypes'] = 'e1/e3'
    #E1/E4
    df.loc[(df['Allele-1']=='APOE1') & (df['Allele-2']=='APOE4'), 'Genotypes'] = 'e1/e4'
    df.loc[(df['Allele-1']=='APOE4') & (df['Allele-2']=='APOE1'), 'Genotypes'] = 'e1/e4'
    #E2/E3
    df.loc[(df['Allele-1']=='APOE2') & (df['Allele-2']=='APOE3'), 'Genotypes'] = 'e2/e3'
    df.loc[(df['Allele-1']=='APOE3') & (df['Allele-2']=='APOE2'), 'Genotypes'] = 'e2/e3'
    #E2/E4
    df.loc[(df['Allele-1']=='APOE2') & (df['Allele-2']=='APOE4'), 'Genotypes'] = 'e2/e4'
    df.loc[(df['Allele-1']=='APOE4') & (df['Allele-2']=='APOE2'), 'Genotypes'] = 'e2/e4'
    #E3/E4
    df.loc[(df['Allele-1']=='APOE3') & (df['Allele-2']=='APOE4'), 'Genotypes'] = 'e3/e4'
    df.loc[(df['Allele-1']=='APOE4') & (df['Allele-2']=='APOE3'), 'Genotypes'] = 'e3/e4'
    
    df.to_csv('apoe_tags.csv', index=False, columns=['eid','rs429358-1', 'rs7412-1', 'Allele-1', 'rs429358-2', 'rs7412-2', 'Allele-2', 'Genotypes']) 

main()

```

- rs429358(C;T) + rs7412 (C;C) = APOE4/APOE3 (Bad)

- rs429358(T;T) + rs7412 (C;T) = APOE3/APOE2 (Good)
- rs429358(T;T) + rs7412 (T;T) = APOE2/APOE2 (Good) -- lowest risk
- rs429358(T;C) + rs7412 (T;C) = APOE2/APOE4 (ambigious)

# Step 2 

```python
data = pd.read_csv('HCDMNCOV_AD_updated_correct_imputation_2022.csv')

e2_scores = list()
for i in range(0,len(data)):
    score = 0
    if (data['Allele-1'][i]=='APOE2'):
        score = score +1 
    if (data['Allele-2'][i]=='APOE2'):
        score = score +1
    #ambiguous e2/e4 
    if (data['Genotypes'][i]=='e1/e3'):
        score = score + 1
    e2_scores.append(score)

data['e2_ranked']=e2_scores

e4_scores = list()
for i in range(0,len(data)):
    score = 0
    if (data['Allele-1'][i]=='APOE4'):
        score = score +1 
    if (data['Allele-2'][i]=='APOE4'):
        score = score +1
    #ambiguous e2/e4 
    if (data['Genotypes'][i]=='e1/e3'):
        score = score + 1 
    e4_scores.append(score)

data['e4_ranked']=e4_scores

all_scores = list()
for i in range(0,len(data)):
    score = 0
    if (data['Allele-1'][i]=='APOE4'):
        score = score -1 
    if (data['Allele-2'][i]=='APOE4'):
        score = score -1
    if (data['Allele-1'][i]=='APOE2'):
        score = score +1 
    if (data['Allele-2'][i]=='APOE2'):
        score = score +1    
    all_scores.append(score)

data['APOE_ranked']=all_scores

data = data.set_index('eid',drop=False)

igf1 = pd.read_csv('IGF1.csv').set_index('userID')

result = pd.concat([data, igf1], axis=1)

result = result.dropna(subset=['eid'])

result = result.rename(columns={"30770": "IGF1"})

result.to_csv('HCDMNCOV_AD_updated_correct_imputation_2022.csv')
```
